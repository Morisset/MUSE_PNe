```python
import numpy as np
import matplotlib.pyplot as plt
import pyneb as pn
```

```python
He1 = pn.RecAtom('He',1)
```

```python
temp = np.linspace(5000, 20000, 1000)

f, ax =plt.subplots()
for dens in (100, 1000, 2000, 5000, 1000, 15000):
    RHe1 = He1.getEmissivity(temp, dens, wave=7281) / He1.getEmissivity(temp, dens, wave=6678)
    ax.scatter(RHe1, temp, s=1)
```

```python
temp = np.linspace(5000, 20000, 100)
dens = np.logspace(2, 6, 200)
RHe1 = He1.getEmissivity(temp, dens, wave=7281) / He1.getEmissivity(temp, dens, wave=6678)
f, ax =plt.subplots()
cs=ax.contour(np.log10(dens), temp, RHe1)
ax.clabel(cs)
ax.set_xlabel('log Ne [cm-3]')
ax.set_ylabel('Te [K]');
```

```python
temp = np.linspace(5000, 20000, 100)
dens = np.logspace(2, 6, 200)
waves = (7065, 7281, 5876, 4713)
f, axes =plt.subplots(2,2,figsize=(12,12))
for wave, ax in zip(waves, axes.ravel()):
    RHe1 = He1.getEmissivity(temp, dens, wave=wave) / He1.getEmissivity(temp, dens, wave=6678)
    cs=ax.contour(np.log10(dens), temp, RHe1)
    ax.clabel(cs)
    ax.set_xlabel('log Ne [cm-3]')
    ax.set_ylabel('Te [K]')
    ax.set_title('{}/6678'.format(wave));
```
